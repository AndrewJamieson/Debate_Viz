```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from tensorflow.keras.preprocessing.text import Tokenizer

import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import *
nltk.download('punkt')
nltk.download('stopwords')

import string
```

```python
df_clean = pd.read_csv('clean.csv')  
```

```python
df_clean.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# df_clean['tokenized'] = df_clean.apply(lambda row: nltk.word_tokenize(row['Txt']), axis=1)
```

```python
df_clean['words'] = df_clean.apply(lambda row: (row['Txt'].split()), axis=1)
```

```python
df_clean['word_count'] = df_clean.apply(lambda row: len(row['words']), axis=1)
```

```python
stopwrd = stopwords.words('english')
```

```python
stop2 =[ 'you’re',
 'i’m',
 "we’re",
 "we're",
 "that’s",
 "that's",
 "i'm",
 "you're",
 '‘i’m']
```

```python
# top 16 words + she,  from Reading Teacher's Book of Lists
# stopwrd = ['the', 'of', 'and', 'a', 'to', 'in', 'is', 'you', 'it', 'he', 'was', 'for', 'on', 'are', 'as', 'with', 'she']
```

```python
# combining stopwords and punctuation and curly quotes
punct = list(string.punctuation + '“”’…–')
stopwrds = stopwrd + punct + stop2
```

```python
df_clean['words_reduced'] =df_clean.apply(lambda row:[w.lower() for w in row['words'] if w.lower() not in stopwrds], axis=1)
```
